### Q4.1.1: Download stock data for Apple, IBM, JP Morgan and the Dow Jones Index. Generate key descriptive statistics, and compute the 1 day returns.

```python
import pandas as pd
import matplotlib.pyplot as plt

apple_df = pd.read_csv("data/AAPL.csv", index_col=0)
ibm_df = pd.read_csv("data/IBM.csv", index_col=0)
jpm_df = pd.read_csv("data/JPM.csv", index_col=0)
dji_df = pd.read_csv("data/DJI.csv", index_col=0)

def get_key_stats(df):
    stats = {
        'mean': df['Close'].mean(),
        'median': df['Close'].median(),
        'std_dev': df['Close'].std(),
        'min': df['Close'].min(),
        'max': df['Close'].max()
    }
    return stats

apple_stats = get_key_stats(apple_df)
ibm_stats = get_key_stats(ibm_df)
jpm_stats = get_key_stats(jpm_df)
dji_stats = get_key_stats(dji_df)

print("Apple Key Stats:")
for key, value in apple_stats.items():
    print(f"  {key.capitalize()}: {value}")
print("")
print("IBM Key Stats:")
for key, value in ibm_stats.items():
    print(f"  {key.capitalize()}: {value}")
print("")
print("JP Morgan Key Stats:")
for key, value in jpm_stats.items():
    print(f"  {key.capitalize()}: {value}")
print("")
print("Dow Jones Index Key Stats:")
for key, value in dji_stats.items():
    print(f"  {key.capitalize()}: {value}")


def compute_daily_returns(df):
    df['Return'] = df['Adj Close'].pct_change()
    return df['Return']

compute_daily_returns(apple_df)
compute_daily_returns(ibm_df)
compute_daily_returns(jpm_df)
compute_daily_returns(dji_df)
```

# 4.2: Robust Estimators

### Q4.2.1 Create functions to find the Median, IQR and MAD from a `panda.series`.

```python
# functinon to find median of panda.series
def median(series):
    sorted_series = series.sort_values()
    n = len(sorted_series)
    if n % 2 == 1:
        return sorted_series.iloc[n // 2]
    else:
        return (sorted_series.iloc[n // 2 - 1] + sorted_series.iloc[n // 2]) / 2

# function to find IQR of panda.series
def iqr(series):
    sorted_series = series.sort_values()
    n = len(sorted_series)
    q1 = sorted_series.iloc[n // 4]
    q3 = sorted_series.iloc[3 * n // 4]
    return q3 - q1

# function to find MAD of panda.series
def mad(series):
    med = median(series)
    return median((series - med).abs())
```

### Q4.2.2: Assess and compare the respective relative computational efficiency of the different estimators.

Computing the median and the interquartile range both require sorting the series and then just finding points at set indexes in the series. The sorting likely takes `O(nlog(n))` and the final index picking takes constant time, so both have a time complexity of `O(nlog(n))`.

The Median Abolute Deviation requires computing the series median then recording the deviation of each element from this median, and then recomputing the median of these deviations. Whilst it still has a time complexity of `O(nlog(n))`, this must compute the median twice making it the most computationally expensive.

### Q4.2.3: Assess and compare the breakdown points of each of the estimators.

The breakpoint of an estimator is the smallest proportion of observations that must be replaced with extreme values to make the estimator go to infinity (or become meaningless).

For the median, this is 50%, after controlling 50% of the series data either side of the median, you can push the median value of the series to any desired value. This is the smallest proportion of the series that must be corrupted in order to have this effect.

The breakdown point of each quartile is 25% for the same reason, so to set the IQR to any desired value you must "breakdown" both quartiles, giving the interquartile range a breakpoint of 50%.

The median absolute deviation also has a breakpoint of 50%, since it means taking the median twice, both with a robustness of 50%.

This can be contrasted with the mean which has a breakdown point of just 0%, as one extreme outlier can destroy it.

# 4.3: Robust and OLS Regression

### Q4.3.1: Regress each stock's 1-day returns against the 1-day returns of the Dow Jones Index using Ordinary Least Squares (OLS) regression.

```python
# regress each stock's 1-day returns against the 1-day returns of the Dow Jones Index using Ordinary Least Squares (OLS) regression
import statsmodels.api as sm

def linear_regression(independent_var, dependent_var_dict):
    results = {}
    X = sm.add_constant(independent_var.dropna())

    for stock, returns in dependent_var_dict.items():
        model = sm.OLS(returns, X).fit()
        results[stock] = model

    # Print the regression results
    for stock, model in results.items():
        print(f"Regression results for {stock}:")
        print(f"Alpha: {model.params['const']}")
        print(f"Beta: {model.params['Return']}")
        print(f"Residual Std Dev: {model.resid.std()}")
        print("\n")

    # Plot the regression lines
    plt.figure(figsize=(12, 4))
    for i, (stock, model) in enumerate(results.items()):
        stock_data = stock_returns[stock].reindex(X.index, fill_value=np.nan).dropna()  # Align stock data with X
        plt.subplot(1, 3, i + 1)
        plt.scatter(dji_df['Return'].dropna(), stock_data, alpha=0.6, label='Data Points')
        plt.plot(dji_df['Return'].dropna(), model.predict(X), color='red', label='OLS Fit')
        plt.title(f'{stock} Returns vs Dow Jones Returns', fontsize=10)
        plt.xlabel('Dow Jones 1 Day Return', fontsize=10)
        plt.ylabel(f'{stock} 1 Day Return', fontsize=10)
        plt.legend()
        plt.grid(False)
    plt.tight_layout()
    plt.show()

    return results

stock_returns = {
    'Apple': apple_df['Return'].dropna(),
    'IBM': ibm_df['Return'].dropna(),
    'JP Morgan': jpm_df['Return'].dropna()
}

stock_returns_outlier = {
    'Apple': apple_outlier_df['Return'].dropna(),
    'IBM': ibm_outlier_df['Return'].dropna(),
    'JP Morgan': jpm_outlier_df['Return'].dropna()
}

print("OLS Regression without Outliers:")
print("")
linear_non_outlier_results = linear_regression(dji_df['Return'], stock_returns)
print("OLS Regression with Outliers:")
print("")
linear_outlier_results = linear_regression(dji_df['Return'], stock_returns_outlier)
```

### Q4.3.2: Regress the 1 day returns of each stock against the 1 day returns of the Dow Jones Index using Huber Regression.

```python
# Regress the 1 day returns of each stock against the 1 day returns of the Dow Jones Index using Huber Regression
from sklearn.linear_model import HuberRegressor

def huber_regression(independent_var, dependent_var_dict):
    results = {}
    X = dji_df['Return'].dropna().values.reshape(-1, 1)

    for stock, returns in dependent_var_dict.items():
        model = HuberRegressor().fit(X, returns)
        results[stock] = model

    # Print the regression results
    for stock, model in results.items():
        print(f"Regression results for {stock}:")
        print(f"Coefficients: {model.coef_}")
        print(f"Intercept: {model.intercept_}")
        print(f"Residual Std Dev: {np.sqrt(np.mean((model.predict(X) - returns) ** 2))}")
        print("\n")

    # Plot the regression lines
    plt.figure(figsize=(12, 4))
    for i, (stock, model) in enumerate(results.items()):
        stock_data = stock_returns[stock].reindex(dji_df['Return'].dropna().index).dropna()  # Align stock data with DJI returns
        aligned_X = dji_df['Return'].dropna().reindex(stock_data.index).values.reshape(-1, 1)  # Align X with stock data
        plt.subplot(1, 3, i + 1)
        plt.scatter(aligned_X, stock_data, alpha=0.6, label='Data Points')
        plt.plot(aligned_X, model.predict(aligned_X), color='red', label='Huber Fit')
        plt.title(f'{stock} Returns vs Dow Jones Returns (Huber)', fontsize=10)
        plt.xlabel('Dow Jones 1 Day Return', fontsize=10)
        plt.ylabel(f'{stock} 1 Day Return', fontsize=10)
        plt.legend()
        plt.grid(False)
    plt.tight_layout()
    plt.show()

    return results

stock_returns = {
    'Apple': apple_df['Return'].dropna(),
    'IBM': ibm_df['Return'].dropna(),
    'JP Morgan': jpm_df['Return'].dropna()
}

stock_returns_outlier = {
    'Apple': apple_outlier_df['Return'].dropna(),
    'IBM': ibm_outlier_df['Return'].dropna(),
    'JP Morgan': jpm_outlier_df['Return'].dropna()
}

print("OLS Regression without Outliers:")
print("")
huber_non_outlier_results = huber_regression(dji_df['Return'], stock_returns)
print("OLS Regression with Outliers:")
print("")
huber_outlier_results = huber_regression(dji_df['Return'], stock_returns_outlier)
```

### Q4.3.3: Assess and compare the results obtained using both regression methods and the impact of outliers in both cases.

```python
# compare the results of the OLS regression and Huber regression in terms of their sensitivity to outliers
print("Comparison of OLS and Huber Regression Results:")
for stock in stock_returns.keys():
    ols_non_outlier = linear_non_outlier_results[stock]
    ols_outlier = linear_outlier_results[stock]
    huber_non_outlier = huber_non_outlier_results[stock]
    huber_outlier = huber_outlier_results[stock]

    print(f"{stock}:")
    print(f"  OLS Non-Outlier Coefficients: {ols_non_outlier.params.values}")
    print(f"  OLS Outlier Coefficients: {ols_outlier.params.values}")
    print(f"  Huber Non-Outlier Coefficients: {huber_non_outlier.coef_}, Intercept: {huber_non_outlier.intercept_}")
    print(f"  Huber Outlier Coefficients: {huber_outlier.coef_}, Intercept: {huber_outlier.intercept_}")
    print("")
```

This side by side comparison of how the regression results change after adding outliers shows that OLS is very sensitive to outliers, whilst Huber regression is much more robust.

# 4.4: Robust Trading Strategies

### Q4.4.1: For each stock, plot the Adjusted Close value, the 20-day dnd 50-day MAs, and highlight crossover periods and the regions to buy or sell. Then repeart with outlier values.

```python
# Plot the adjusted close value, the 20-day and 50-day MAs, and highlight crossover periods and the regions to buy or sell for each stock, then repeat with outlier values
def plot_moving_averages(df, title, color, ax, average_method="mean"):
    if average_method == "mean":
        df['MA20'] = df['Adj Close'].rolling(window=20).mean()
        df['MA50'] = df['Adj Close'].rolling(window=50).mean()
    elif average_method == "median":
        df['MA20'] = df['Adj Close'].rolling(window=20).median()
        df['MA50'] = df['Adj Close'].rolling(window=50).median()

    ax.plot(df['Adj Close'], label=f'{title} Adjusted Close', color=color)
    ax.plot(df['MA20'], label=f'{title} 20-Day MA', color='cyan')
    ax.plot(df['MA50'], label=f'{title} 50-Day MA', color='magenta')

    # Highlight crossover periods
    crossover_buy = (df['MA20'] > df['MA50']) & (df['MA20'].shift(1) <= df['MA50'].shift(1))
    crossover_sell = (df['MA20'] < df['MA50']) & (df['MA20'].shift(1) >= df['MA50'].shift(1))

    ax.scatter(df.index[crossover_buy], df['Adj Close'][crossover_buy], marker='^', color='green', label='Buy Signal', s=100)
    ax.scatter(df.index[crossover_sell], df['Adj Close'][crossover_sell], marker='v', color='red', label='Sell Signal', s=100)

    ax.set_title(f'{title} Adjusted Close and Moving Averages', fontsize=12)
    ax.set_xlabel('Date', fontsize=10)
    ax.set_ylabel('Price', fontsize=10)
    ax.legend()
    ax.xaxis.set_major_locator(plt.MaxNLocator(X_LABELS))  # Limit the number of x-axis labels

    # Calculate returns based on buy and sell signals
    positions = pd.Series(index=df.index, data=0)  # Initialize positions
    positions[crossover_buy] = 1  # Buy signal
    positions[crossover_sell] = -1  # Sell signal
    positions = positions.ffill().fillna(0)  # Forward fill positions

    returns = positions.shift(1) * df['Return']  # Calculate returns based on positions
    cumulative_returns = (1 + returns).cumprod() - 1  # Cumulative returns

    return cumulative_returns


plt.figure(figsize=(12, 12))
apple_ret = plot_moving_averages(apple_df, 'Apple', 'blue', plt.subplot(2, 2, 1))
ibm_ret = plot_moving_averages(ibm_df, 'IBM', 'orange', plt.subplot(2, 2, 2))
jpm_ret = plot_moving_averages(jpm_df, 'JP Morgan', 'green', plt.subplot(2, 2, 3))
dji_ret = plot_moving_averages(dji_df, 'Dow Jones Index', 'red', plt.subplot(2, 2, 4))
plt.tight_layout()
plt.show()  

plt.figure(figsize=(12, 12))
apple_outlier_ret = plot_moving_averages(apple_outlier_df, 'Apple', 'blue', plt.subplot(2, 2, 1))
ibm_outlier_ret = plot_moving_averages(ibm_outlier_df, 'IBM', 'orange', plt.subplot(2, 2, 2))
jpm_outlier_ret = plot_moving_averages(jpm_outlier_df, 'JP Morgan', 'green', plt.subplot(2, 2, 3))
dji_outlier_ret = plot_moving_averages(dji_outlier_df, 'Dow Jones Index', 'red', plt.subplot(2, 2, 4))
plt.tight_layout()
plt.show()  

print("Cumulative Returns without Outliers:")
print(f"Apple: {apple_ret.iloc[-1]:.2%}")
print(f"IBM: {ibm_ret.iloc[-1]:.2%}")
print(f"JP Morgan: {jpm_ret.iloc[-1]:.2%}")
print(f"Dow Jones Index: {dji_ret.iloc[-1]:.2%}")

print("\nCumulative Returns with Outliers:")
print(f"Apple: {apple_outlier_ret.iloc[-1]:.2%}")
print(f"IBM: {ibm_outlier_ret.iloc[-1]:.2%}")
print(f"JP Morgan: {jpm_outlier_ret.iloc[-1]:.2%}")
print(f"Dow Jones Index: {dji_outlier_ret.iloc[-1]:.2%}")
```
